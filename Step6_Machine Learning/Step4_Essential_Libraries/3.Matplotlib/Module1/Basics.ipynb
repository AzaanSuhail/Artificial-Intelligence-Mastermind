## Types of Graphs

### 1. Error Graph
An error bar graph shows:
* A central value (mean / measurement)
* A range of uncertainty or variation around it

Common use cases:
* Experimental results
* Measurement uncertainty
* Standard deviation / standard error
* Confidence intervals
  


Type os Error Graph : 
  * Horizontal Error Bars
  * Vertical Error Bars

```python
# Creating a Simple Graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 1, 2, 1, 2, 1]

plt.plot(x, y)
```

### 1. Adding Error to the y-axis

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 1, 2, 1, 2, 1]

# creating error
y_error = 0.2

# plotting graph
plt.plot(x, y) #* Normal graph

plt.errorbar(x, y, yerr=y_error, fmt="o")
```

### 2. Adding Error to the x-values 

```python
# creating error
x_error = 0.5

# plotting graph
plt.plot(x, y)
plt.errorbar(x, y, xerr=x_error, fmt="o")
```

### 3. Adding Error to both x & y

```python
# creating error
x_error = 0.5
y_error = 0.3

# plotting graph
plt.plot(x, y)
plt.errorbar(x, y, yerr=y_error, xerr=x_error, fmt="o")
```

### 4. Variable Error in x & y

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 1, 2, 1]

# creating error
y_errormin = [0.1, 0.5, 0.9, 0.1, 0.9]
y_errormax = [0.2, 0.4, 0.6, 0.4, 0.2]

x_error = 0.5
y_error = [y_errormin, y_errormax]

# plotting graph
# plt.plot(x, y)
plt.errorbar(x, y, yerr=y_error, xerr=x_error, fmt="o")
```

### 2. Line Chart


```python
import matplotlib.pyplot as plt
import numpy as np

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
temperature = [22, 24, 23, 26, 25]

plt.plot(days, temperature, marker="o")
plt.title("Weekly Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.show()
```

### Simple Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = [3, 6, 9, 12, 15]

plt.plot(x, y, marker="o", linestyle="-", label="Data Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Markers")
plt.legend()
plt.show()
```

### Adding Grid to a Line Chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.grid(True) #LEARN
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Grid")
plt.show()
```

### Line Char with Annotations 
* It display additional info such as the exact x & y directly to the data points

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker="o", linestyle="-")

for xi, yi in zip(x, y):
    plt.annotate(
        f"({xi}, {yi})",
        (xi, yi),
        textcoords="offset points",
        xytext=(0, 10),
        ha="center",
    )

plt.title("Line Chart with Annotations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### Fill the Area Between Two Lines

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2
y1 = [3, 5, 7, 9]

plt.plot(x, y, label="y = 2x")
plt.plot(x, y1, "-.", label="y1")

plt.fill_between(x, y, y1, color="green", alpha=0.4)

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title("Filled Area Between Two Lines")
plt.legend()
plt.show()
```

### Saving a Line Chart to a File
