## What is Pandas ?
* Pandas (stands for Python Data Analysis) is an open-source software library designed for data manipulation and analysis.
* Revolves around two primary **Data structures**: Series (1D) and DataFrame (2D).
* Built on top of NumPy, efficiently manages large datasets, offering tools for data cleaning, transformation, and analysis.



### Data Frame
* It is a two-dimensional data structure constructed with rows and columns, which is more similar to Excel spreadsheet.


```python
import pandas as pd
```

```python
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s

print("--------------")

s = pd.Series()
print("Pandas Series: ", s)

# NOTE : In pandas we don't labeled it, it will labelled automatically from 0,1,2,3,4 ....
```

### 3. Handling Missing Data

```python
print(df.isnull().sum())
df = df.fillna(0)
```

### 4. Selecting and Filtering Data

```python
ages = df[df["age"] > 25]
print(ages)
```
